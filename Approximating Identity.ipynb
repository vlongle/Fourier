```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
eps = 0.1
p = lambda theta: eps + np.cos(theta)
```

```python
thetas = np.linspace(-2*np.pi, 2*np.pi, 100)

p30 = p(thetas) ** 30
p15 = p(thetas) ** 15
p6 = p(thetas) ** 6
p1 = p(thetas) 
```

```python
plt.plot(thetas, p30, label='p30')
plt.plot(thetas, p15, label='p15')
plt.plot(thetas, p6, label='p6')
plt.plot(thetas, p1, label='p1')
plt.legend()
# region of interest
plt.axvline(x=np.pi, linestyle='--', color='black')
plt.axvline(x=-np.pi, linestyle='--', color='black')
```

```python
# pw bounded but not uniformly.
# f_n(x)=1/(nx) on (0, 1)

xs = np.linspace(0.01, 1, 100)
fn = lambda x: 1/x
f1 = fn(xs)/1
f5 = fn(xs)/5
f20 = fn(xs)/20
```

```python
plt.plot(xs, f1, label='f1')
plt.plot(xs, f5, label='f2')
plt.plot(xs, f20, label='f3')
plt.legend()
```
