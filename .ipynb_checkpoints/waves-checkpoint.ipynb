## Standing Wave Harmonic

```python
dx=0.1 #space increment
dt=0.05 #time increment
tmin=0.0 #initial time
tmax=2.0 #simulate until
xmin=- 2 * np.pi #left bound
xmax= 2 * np.pi#right bound...assume packet never reaches boundary

# Not needed probably.
c=1.0 #speed of sound. In the wave eqn
rsq=(c*dt/dx)**2 #appears in finite diff sol

nx = int((xmax-xmin)/dx) + 1 #number of points on x grid
nt = int((tmax-tmin)/dt) + 2 #number of points on t grid
u1 = np.zeros((nt,nx)) #solution to WE
u2 = np.zeros((nt,nx)) #solution to WE



for a in range(0,nx):
    x = xmin + a * dx
    u1[0,a]=np.sin(x)
    u2[0,a]= np.sin(3 * x)


xs = xmin + dx * np.array(range(0, nx))
#simulate dynamics
for t in range(0,nt-1):
    for a in range(0,nx):
        x = xmin + a * dx
        u1[t+1,a] = np.cos(t * dt) * np.sin(x) # fundamental tone
        u2[t+1,a] = np.cos(3 * t * dt) * np.sin( 3 * x) # overtone
```

```python
# https://en.wikipedia.org/wiki/Wave_equation#Investigation_by_numerical_methods
p = 1.5
h = 1 
c= 1 
tmax=2.0 #simulate until
xmin= 0 #left bound
xmax= np.pi#right bound...assume packet never reaches boundary

nx = int((xmax-xmin)/dx) + 1 #number of points on x grid
nt = int((tmax-tmin)/dt) + 2 #number of points on t grid
```

```python
# initial profile: plucked string
def init_fn(x):
    if x < p:
        return (x * h)/p
    else:
        return   (h * (np.pi - x))/(np.pi - p)
```

```python
# wave eqn: u_tt = c u_xx
# http://hplgit.github.io/num-methods-for-PDEs/doc/pub/wave/html/slides_wave-solarized.html


u = np.zeros((nt,nx)) #solution to WE

xs = xmin + dx * np.array(range(0, nx))

for a in range(0,nx):
    u[0,a]=init_fn(xmin+a*dx)
    u[1,a]=u[0,a]

    
# http://www.physics.purdue.edu/~giordano/tmp/book_strings/book_strings.html

r = (c * dt)/dx
for t in range(2,nt-1):
    for a in range(1,nx-1):
        # laplacian
        lap = u[t, a+1] - 2 * u[t, a] + u[t, a-1]
        u[t+1,a] = - u[t-1, a] + 2 * u[t, a] + (c ** 2) * lap
#        
#        u[t+1, a] = 2 * (1-r*r)*u[t, a] - u[t-1, a] + r*r*(
#                    u[t, a+1] + u[t, a-1])
```

## Traveling wave

```python
tmax=8.0 #simulate until
xmin= -2*np.pi #left bound
xmax= 2 * np.pi#right bound...assume packet never reaches boundary

l = -2 * np.pi +(1/3)
h = -1/3

c = 1 # speed of the moving wave

nx = int((xmax-xmin)/dx) + 1 #number of points on x grid
nt = int((tmax-tmin)/dt) + 2 #number of points on t grid
```

```python
u = np.zeros((nt,nx)) #solution to WE
```

```python
def F(x):
    if x < h and x > l:
        return np.sin(x)
    else:
        return 0 
```

```python
# u(x, 0) = F(x)
for a in range(0,nx):
    u[0,a] = F(xmin+a*dx)

    
xs = xmin + dx * np.array(range(0, nx))

# u(x, t) = F(x - c * t), moving to the right!
for t in range(1,nt):
    for a in range(0,nx):
        u[t, a] =  F(xs[a] - c * t * dt)
        
```

```python
plt.plot(u[0, :])
```

```python
fig = plt.figure()
plts = []             # get ready to populate this list the Line artists to be plotted
for i in range(0, nt):
    p, = plt.plot( xs ,u[i,:], 'r', alpha=0.8)   # this is how you'd plot a single line...
    plts.append( [p] )           # ... but save the line artist for the animation
ani = animation.ArtistAnimation(fig, plts, interval=100, # interval =
                                # speed of the simulation
                                repeat_delay=3000, blit=True)   # run the animation
ani.save('traveling_wave.mp4')    # optionally save it to a file
```
