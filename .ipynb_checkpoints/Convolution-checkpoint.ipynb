```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm #  normal dist
```

```python
xmin, xmax = -2*np.pi, 2*np.pi
dx = 0.01
nx = 100
xs = np.linspace(xmin, xmax, nx)

L, H = 1, 4
square = lambda x: x if x > L and x < H else 0
ys = np.zeros(nx)


# F(x)
for i in range(nx):
    ys[i] = square(xs[i])

# F(-x)
ys_f = np.zeros(nx)
for i in range(nx):
    ys_f[i] = square(-xs[i])

# F(pi-x). Kernel for conv
# (F * g)(pi)

ys_k = np.zeros(nx)
for i in range(nx):
    ys_k[i] = square(np.pi-xs[i])
```

```python
plt.plot(xs, ys)
plt.plot(xs, ys_f, '--')
plt.plot(xs, ys_k)

#plt.axvline(x=np.pi, linestyle='--', color='black')
#plt.axvline(x=0, linestyle='--', color='black')

# green graph is the orange graph shifted by pi
```

```python
gaussian = lambda x: norm.pdf(x, 0, 1) # decrease variance
# to make the kernel more concentrated ...


ys = np.zeros(nx)


# F(x)
for i in range(nx):
    ys[i] = gaussian(xs[i])

# F(-x)
ys_f = np.zeros(nx)
for i in range(nx):
    ys_f[i] = gaussian(-xs[i])

# F(pi-x). Kernel for conv
# (F * g)(pi)

ys_k = np.zeros(nx)
for i in range(nx):
    ys_k[i] = gaussian(np.pi-xs[i])
    
```

```python
plt.plot(xs, ys)
plt.plot(xs, ys_f, '--')
plt.plot(xs, ys_k)

plt.axvline(x=np.pi, linestyle='--', color='black')
```
